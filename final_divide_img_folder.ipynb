```python
def status_to_int(status):
    if status == 'NOR': #정상
        int_status = 0
    else:
        int_status = 1
    return int_status
```

```python
def json_name_to_img_path(json_name):
    img_name = json_name.split('.')[0] +'.jpg'
    img_path = img_dir_path + '/'+ img_name
    return img_path
```

```python
def padding(img_path, annotation, train_val):
    img = Image.open(img_path).convert('1')
    img = img.convert('RGB')
    
    w, h = img.size

    img = np.array(img)

    margin = [np.abs(w - h) // 2 , np.abs(w - h) // 2]
    if w > h:
        margin_list = [margin, [0,0]]
        w_plus = 0
        h_plus = margin[0]
    else:
        margin_list = [[0,0], margin]
        w_plus = margin[0]
        h_plus = 0
        

    if len(img.shape) == 3:
        margin_list.append([0,0])

    output = np.pad(img, margin_list, mode='constant')
    img = Image.fromarray(output)
        
    if train_val =='t':
        [[x,y], [w,h]] =  annotation

        # 결측치 조절
        if x> w:
            x, w = w,x

        if y>h:
            y, h = h,y
            
        x +=w_plus
        w +=w_plus
        y +=h_plus
        h +=h_plus

        xx = img.size[0]/7
        yy = img.size[1]/7  

        if xx<x and yy<y and w<img.size[0]-xx and h<img.size[1]-yy:   # box가 범위보다 작음
            cropped_img = img.crop((xx, yy, img.size[0]-xx, img.size[1]-yy))
        else:
            if xx>=x:
                xx=x
                ww=img.size[0]-xx
                hh=img.size[1]-yy
            if yy>=y:
                yy=y
                ww=img.size[0]-xx
                hh=img.size[1]-yy

            if w>=img.size[0]-xx:
                ww = w
                hh = img.size[1]-yy

            if h>=img.size[1]-yy:
                hh = h
                ww=img.size[0]-xx
            cropped_img = img.crop((xx, yy, ww, hh))
        img = cropped_img


        w, h = img.size

        img = np.array(img)

        margin = [np.abs(w - h) // 2 , np.abs(w - h) // 2]
        if w > h:
            margin_list = [margin, [0,0]]
        else:
            margin_list = [[0,0], margin]

        if len(img.shape) == 3:
            margin_list.append([0,0])

        output = np.pad(img, margin_list, mode='constant')
        img = Image.fromarray(output)

    
    # 이미지 크기
    img = img.resize((224, 224))
    #img = np.array(img)
    return img

```
